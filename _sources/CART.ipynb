# Materi 6 : CART (classification and regression tree)

##Menggunakan data iris

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
print(iris.target_names)
print(iris.feature_names)
data=iris.data
data.shape
# 0 - 49 = setosa
# 50 - 99 = versicolor
# 100 - 149 = virginica
```

```python
#dua baris pertama dan fitur index ke - 0
fitur1=data[0:2,:2]
fitur2=data[50:52,:2]
fitur3=data[100:102,:2]
fitur=np.concatenate((fitur1,fitur2,fitur3))
print(fitur)
```

```python
target = np.array([0,0,1,1,2,2])
fitur = np.reshape(fitur,(6,2)) 
print(fitur)
```

### Training Data

```python
treeClf = DecisionTreeClassifier()
treeClf.fit(fitur,target)
```

### Testing Data

```python
print(treeClf.predict(fitur))
```

### Display Tree

#### Menggunakan sklearn tree classification

```python
from sklearn import tree
```

```python
tree.plot_tree(treeClf)
```

#### Menggunakan graphviz

```python
import graphviz 
dot_data = tree.export_graphviz(treeClf, out_file=None,feature_names=iris.feature_names[0:2],class_names=iris.target_names) 
graph = graphviz.Source(dot_data) 
# graph.render("iris") 
graph
```

#### Menggunakan tree regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
target = np.array([10,14,23,25,31])
fitur = np.array([[2],[3],[4],[5],[6]]) 
```

```python
treeReg = DecisionTreeRegressor(max_depth=2)
treeReg.fit(fitur,target)
```

```python
import graphviz 
dot_data = tree.export_graphviz(treeReg, out_file=None,feature_names=fitur[0:1],class_names=target) 
graph = graphviz.Source(dot_data) 
# graph.render("iris") 
graph
```

```python
print(treeReg.predict([[6]]))
```
