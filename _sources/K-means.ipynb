# K-Means

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data

```python
data = np.array([[7,5],[5,7],[7,7],[3,3],[4,6],[1,4]])
data.shape
```

## Perhitungan Manual

```python
def updateCentroid(dataTraining,minDist,k):
  centroid = np.zeros((k,dataTraining.shape[1]))
  for i in range(k):
    centroid[i]=np.sum(dataTraining[minDist==i],axis=0)/np.sum(minDist==i) #minDist = dist yang memilik cluster yang sama
  return centroid

def trainKmeans(dataTraining,k,epoch):
  centroid = np.array([[5,7],[7,7]]) #menentukan centroid awal
  jmlhData = dataTraining.shape[0]
  for ap in range(epoch):
    dist=np.zeros((jmlhData,k)) #jmlhdata * jmlhkelas
    cluster = np.zeros((jmlhData))
    for i in range(jmlhData):
      for j in range(k):
        dist[i,j] = np.sqrt(np.sum((data[i]-centroid[j])**2)) #ecludian distance
      cluster[i]=np.argmin(dist[i])
    centroid = updateCentroid(dataTraining,cluster,k)
    # print(centroid)
  return centroid
```

```python
centroid = trainKmeans(data,2,2)
```

```python
plt.scatter(data[:,0],data[:,1])
plt.scatter(centroid[0,0],centroid[0,1],color='r')
plt.scatter(centroid[1,0],centroid[1,1],color='g')
```
