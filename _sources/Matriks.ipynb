# Materi 1 : Matriks Dasar

```python
import numpy as np
import matplotlib.pyplot as matplot
from sklearn import datasets
%matplotlib inline
```

## Matriks 1 Dimensi
sekelompok data yang disusun dalam satu baris atau satu kolom, yang dapat direpresentasikan dalam bentuk array satu dimensi.

```python
data = np.array([3,5,6])
print(data)
shape = data.shape
print('Ukuran Matriks =',shape)
```

## Matriks 2 Dimensi
sekelompok data yang disusun dalam beberapa baris dan kolom, yang dapat direpresentasikan dalam bentuk array dua dimensi.

```python
data2 = np.array([[3,5,6],[1,3,2]])
print(data2)
# cek jumlah baris dan kolom
shape = data2.shape
print('Ukuran Matriks =',shape)
```

## Membuat isi matriks dengan data = 0
### np.zeros((baris,kolom))

```python
zerodata = np.zeros((2,3))
print(zerodata)
```

## Membuat isi matriks dengan data = 1
### np.ones((baris,kolom))

```python
onedata = np.ones((2,3))
print(onedata)
```

## Membuat isi matriks dengan data Terurut
### np.linspace((awal,akhir,jumlahdata))

```python
linspacedata = np.linspace(0,1,10)
print(linspacedata)
```

## Membuat isi matriks dengan data Random
### np.random.rand((baris,kolom))

```python
randomdata = np.random.rand(2,3)
print(randomdata)
```

## Operasi Element Wise (Operasi Antar Element)

### Penjumlahan

```python
data = np.array([3,5,6])
data2 = np.array([[3,5,6],[1,3,2]])
jumlah = data+data2
print("penjumlahan =")
print(jumlah)
```

### Perkalian

```python
data = np.array([3,5,6])
data2 = np.array([[3,5,6],[1,3,2]])
kali = data*data2
print("Perkalian =")
print(kali)
```

## Operasi Matriks Wise
- Jumlah kolom harus sama dengan baris
- Jika jumlah kolom dan baris sama maka akan error

### Perkalian

```python
data3 = np.random.rand(3,1)
data2 = np.array([[3,5,6],[1,3,2]])
MatriksKali = data2.dot(data3)
print(MatriksKali)
```
