# Summary Model

Output dari klasifikasi adalah output dari layer ini (FC)
Input dari FC layer adalah array 1D

```python
import keras
```

## Fully Connected Layer (Dense)

### Membuat Model
Sequential = beberapa layer

```python
FClayers = keras.models.Sequential()
```

### Menentukan Dense
- Jumlah Bobot/param = (Jmlh Neuron Layer Sebelumnya + 1) * Jmlh Current Layer

```python
# Jumlah Neuron Prev Layer = 6, Jmlh Current Layer = 10
FClayers.add(keras.layers.Dense(10,activation='relu',input_shape=(6,)))
# Jumlah Neuron Prev Layer = 10 (dari dense sebelumnya), Jmlh Current Layer = 3
FClayers.add(keras.layers.Dense(3,activation='relu'))
```

### Menampilkan Jumlah Bobot / Param

```python
FClayers.summary()
```
