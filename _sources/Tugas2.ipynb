```python
import numpy as np
```

```python
class LinearRegressionSGD:
    def __init__(self, learningRate=0.01, epochs=1000, batch_size=32):
        self.learningRate = learningRate
        self.epochs = epochs
        self.batch_size = batch_size
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        n_batches = n_samples // self.batch_size
        
        for _ in range(self.epochs):
            for batch_idx in range(n_batches):
                X_batch = X[batch_idx*self.batch_size:(batch_idx+1)*self.batch_size]
                y_batch = y[batch_idx*self.batch_size:(batch_idx+1)*self.batch_size]
                y_pred = self.predict(X_batch)
                dw = (1/self.batch_size) * np.dot(X_batch.T, (y_pred - y_batch))
                db = (1/self.batch_size) * np.sum(y_pred - y_batch)
                self.weights -= self.learningRate * dw
                self.bias -= self.learningRate * db

    def bias(self):
        return self.bias

    def weights(self):
        return self.weights
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
# membuat data dummy
X = np.array([[1,2], [3,4], [5,6]])
y = np.array([3, 7, 11])

# membuat objek model
model = LinearRegressionSGD(learningRate=0.1, epochs=1000, batch_size=2)

# melatih model
model.fit(X, y)

# melakukan prediksi
X_test = np.array([[7, 8], [9, 10]])
predictions = model.predict(X)

print(predictions)
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# generate dataset
X, y = make_regression(n_samples=1000, n_features=5, noise=20, random_state=1)

# split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# train model
model = LinearRegressionSGD(learningRate=0.01, epochs=1000)
model.fit(X_train, y_train)

# test model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("MSE: ", mse)
```
