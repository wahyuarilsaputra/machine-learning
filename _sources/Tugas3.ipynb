## Aktivasi StepWise

```python
class PerceptronStepWise:
    def __init__(self, learning_rate=0.01, epoch=100):
        self.learning_rate = learning_rate
        self.epoch = epoch
        self.bias = None
        self.bobot = None
    def fit(self, X, y):
        n_data, n_fitur = X.shape
        self.bobot = np.zeros(n_fitur)
        self.bias = 0

        for iter in range(self.epoch):
            # print('epoch-', iter)
            for i in range(n_data):
                netIn = np.dot(X[i], self.bobot) + self.bias
                out = self.aktivasiSigmoid(netIn)
                if out == y[i]:
                  self.bobot = self.bobot
                  self.bias = self.bias
                elif out < y[i]:
                  self.bobot += self.learning_rate*X[i]
                  self.bias += self.learning_rate
                else:
                  self.bobot -= self.learning_rate*X[i]
                  self.bias -= self.learning_rate
            # print(self.bias)
            # print(self.bobot)

    def aktivasi(self, netIn):
        if netIn >= 0 :
          return 1
        else:
          return 0

    def aktivasiSigmoid(self, netIn):
        sig = 1/(1+np.exp(-netIn))
        if sig >= 0.5:
          return 1
        else:
          return 0

    def predict(self, X):
        output = []
        for x in X:
          netIn = np.dot(x, self.bobot) + self.bias
          out = self.aktivasiSigmoid(netIn)
          output.append(out)
        
        return output
```

## Aktivasi Sigmoid

```python
class PerceptronSigmoid:
    def __init__(self, learning_rate=0.02, epoch=1):
        self.learning_rate = learning_rate
        self.epoch = epoch
        self.bias = None
        self.bobot = None
    def fit(self, X, y):
        n_data, n_fitur = X.shape
        # inisialisasi random untuk bias dan coefficient
        self.bobot = np.zeros(n_fitur)
        self.bias = 0

        for iter in range(self.epoch):
            # print('epoch-', iter)
            for i in range(n_data):
                netIn = np.dot(X[i], self.bobot) + self.bias
                out = self.aktivasiSigmoid(netIn)
                # update bobot
                eror = y[i]-out
                self.bobot += self.learning_rate*eror*X[i]
                self.bias += self.learning_rate*eror

                # print('data ke-', i)
            # print(self.bias)
            # print(self.bobot)
          
    def aktivasi(self, netIn):
        if netIn >= 0 :
          return 1
        else:
          return 0
    def aktivasiSigmoid(self, netIn):
        sig = 1/(1+np.exp(-netIn))
        if sig >= 0.5:
          return 1
        else:
          return 0

    def predict(self, X):
        output = []
        for x in X:
          netIn = np.dot(x, self.bobot) + self.bias
          out = self.aktivasiSigmoid(netIn)

          output.append(out)
        
        return output
```

```python
X = np.array([[1,1],[1,0],[0,1],[0,0]])
y_and = np.array([1,0,0,0])
y_Or=np.array([1,1,1,0])
y_notAnd=np.array([0,1,1,1])
y_notOr = np.array([0,1,1,0])

model_StepStepWise = PerceptronStepWise(learning_rate=0.01, epoch=10)
model_Sigmoid = PerceptronSigmoid(learning_rate=0.01, epoch=10)

model_StepStepWise.fit(X,y_Or)
model_Sigmoid.fit(X,y_Or)

y_predStepWise = model_StepStepWise.predict(X)
print('StepWise')
print('Bobot StepWise = ', model_StepStepWise.bobot)
print('Bias StepWise = ', model_StepStepWise.bias)
print('Prediksi Ste Wise = ', y_predStepWise)

print('')
print('Sigmoid')
y_predSigmoid = model_Sigmoid.predict(X)
print('Bobot Sigmoid = ', model_Sigmoid.bobot)
print('Bias Sigmoid = ', model_Sigmoid.bias)
print('Prediksi Sigmoid = ', y_predSigmoid)
```
